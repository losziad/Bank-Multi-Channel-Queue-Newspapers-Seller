```python
import numpy as np
from numpy import random
import matplotlib.pyplot as plt
from prettytable import PrettyTable
```

```python
def generate_vip_interarrival_time(r1):
    if r1<11:
        vip_interarrival = 1
    elif r1<30:
        vip_interarrival = 2
    elif r1<61:
        vip_interarrival = 3
    else:
        vip_interarrival = 4
    return vip_interarrival
```

```python
def generate_vip_service_time(r2):
    if r2<11:
        vip_service_time = 1
    elif r2<41:
        vip_service_time = 2
    elif r2<79:
        vip_service_time = 3
    else:
        vip_service_time = 4
    return vip_service_time
```

```python
def generate_ord_interarrival_time(r3):
    if r3<10:
        ord_interarrival_time = 0
    elif r3<27:
        ord_interarrival_time = 1
    elif r3<54:
        ord_interarrival_time = 2
    elif r3<74:
        ord_interarrival_time = 3
    elif r3<89:
        ord_interarrival_time = 4
    else:
        ord_interarrival_time = 5
    return ord_interarrival_time
```

```python
def generate_ord_service_time(r4):
    if r4<21:
        ord_service_time = 1
    elif r4<61:
        ord_service_time = 2
    elif r4<89:
        ord_service_time = 3
    else:
        ord_service_time = 4
    return ord_service_time
```

```python
num_of_cust = 30
counter = 1000
for i in range(counter):
   vip_interarrival_time_queue.clear()
   vip_service_time_queue.clear()
   ord_arrival_time.clear()
   vip_arrival_time.clear()
   ord_service_start_time.clear()
   vip_service_start_time.clear()
   ord_waiting_time.clear()
   vip_waiting_time.clear()
   vip_comp_time.clear()
   ord_comp_time.clear()
   t_i_s_vip.clear()
   t_i_s_ord.clear()
   vip_idle_time_list.clear()
   ord_idle_time_list.clear()
    
   #vip teller condition 
   for j in range(num_of_cust):
      r1 = random.randint(100)           # v is vip 
      vip_random_number_interarrival_time.append(r1)
      vip_IAT = generate_vip_interarrival_time(r1)
      vip_interarrival_time_queue.append(vip_IAT)
      r2 = random.randint(100)           # v is vip 
      vip_random_number_service_time.append(r2)
      vip_ST = generate_vip_service_time(r2)
      vip_service_time_queue.append(vip_ST)
      if j == 0:
        arrival_time = vip_interarrival_time_queue[j]
        s_s_t = arrival_time
        idle_time = s_s_t
      else:
        arrival_time = vip_arrival_time[j-1] + vip_interarrival_time_queue[j]
        if vip_comp_time[j-1] > arrival_time:
          s_s_t = vip_comp_time[j-1]
        else:
          s_s_t = arrival_time
        idle_time = s_s_t - vip_comp_time[j-1]
      vip_arrival_time.append(arrival_time)
      vip_service_start_time.append(s_s_t)
      w_t = s_s_t - arrival_time
      c_t = vip_service_time_queue[j] + s_s_t
      T_i_s = w_t + vip_service_time_queue[j]
      vip_waiting_time.append(w_t)
      vip_comp_time.append(c_t)
      t_i_s_vip.append(T_i_s)
      vip_idle_time_list.append(idle_time)
    
   for k in range(num_of_cust):
      r3 = random.randint(100)          # o is ordinary 
      ord_random_number_interarrival_time.append(r3)
      ord_IAT = generate_ord_interarrival_time(r3)
      ord_interarrival_time_queue.append(ord_IAT)
      r4 = random.randint(100)         # o is ordinary 
      ord_random_number_service_time.append(r4)
      ord_ST = generate_ord_service_time(r4) 
      ord_service_time_queue.append(ord_ST)
      if k == 0:
        arrival_time = ord_interarrival_time_queue[k]
        s_s_t = arrival_time
        idle_time = s_s_t
      else:
        arrival_time = ord_arrival_time[k-1] + ord_interarrival_time_queue[k]
        if ord_comp_time[k-1] > arrival_time:
          s_s_t = ord_comp_time[k-1]
        else:
          s_s_t = arrival_time
        idle_time = s_s_t - ord_comp_time[k-1]
      ord_arrival_time.append(arrival_time)
      ord_service_start_time.append(s_s_t)
      w_t = s_s_t - arrival_time
      c_t = ord_service_time_queue[k] + s_s_t
      T_i_s = w_t + ord_service_time_queue[k]
      ord_waiting_time.append(w_t)
      ord_comp_time.append(c_t)
      t_i_s_ord.append(T_i_s)
      ord_idle_time_list.append(idle_time)
   average_wait_time_ord = round(sum(ord_waiting_time)/len(ord_waiting_time),2)
   average_wait_time_ord_list.append(average_wait_time_ord)
   average_wait_time_vip = round(sum(vip_waiting_time)/len(vip_waiting_time),2)
   avarege_wait_time_vip_list.append(average_wait_time_vip)
   vip_average_idle_time = round(sum(vip_idle_time_list)/len(vip_idle_time_list),2)
   vip_avarege_idle_time_list.append(vip_average_idle_time)
   ord_average_idle_time = round(sum(ord_idle_time_list)/len(ord_idle_time_list),2)
   ord_avarege_idle_time_list.append(ord_average_idle_time)
   avarege_comp_time_vip = round(sum(vip_comp_time)/len(vip_comp_time),2)
   avarege_comp_time_vip_list.append(avarege_comp_time_vip)
   average_comp_time_ord = round(sum(ord_comp_time)/len(ord_comp_time),2)
   average_comp_time_ord_list.append(average_comp_time_ord)
   avarege_arrival_time_vip = round(sum(vip_arrival_time)/len(vip_arrival_time),2)
   avarege_arrival_time_vip_list.append(avarege_arrival_time_vip)  
   average_arrival_time_ord = round(sum(ord_arrival_time)/len(ord_arrival_time),2)
   average_arrival_time_ord_list.append(average_arrival_time_ord)
```

```python
def calc_max_num_in_ord_queue(count,c,index):
    counter = c
    i = index
    m_count = count
    if(counter < len(ord_arrival_time) and ord_arrival_time[counter]<ord_service_start_time[i]):
        counter = counter + 1
        m_count = m_count + 1
        calc_max_num_in_ord_queue(m_count,counter,i)
    else:
        max_num_stay_in_ord_queue.append(m_count)

for i in range(len(ord_arrival_time)):
    max_count = 0
    count = i
    calc_max_num_in_ord_queue(max_count,count,i)

def calc_max_num_in_vip_queue(count,c,index):
    counter = c
    i = index
    m_count = count
    if(counter < len(vip_arrival_time) and vip_arrival_time[counter]<vip_service_start_time[i]):
        counter = counter + 1
        m_count = m_count + 1
        calc_max_num_in_vip_queue(m_count,counter,i)
    else:
        max_num_stay_in_vip_queue.append(m_count)

for i in range(len(vip_arrival_time)):
    max_count = 0
    count = i
    calc_max_num_in_vip_queue(max_count,count,i)
```
