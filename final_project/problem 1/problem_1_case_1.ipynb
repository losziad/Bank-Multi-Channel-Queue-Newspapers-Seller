```python
import numpy as np
from numpy import random
import matplotlib.pyplot as plt
from prettytable import PrettyTable
```

```python
#this function used to genrate the ( interarrival time )  for the vip customer 

def generate_vip_interarrival_time(r1):
    if r1<11:
        vip_interarrival = 1
    elif r1<30:
        vip_interarrival = 2
    elif r1<61:
        vip_interarrival = 3
    else:
        vip_interarrival = 4
    return vip_interarrival
```

```python
#function to generate the vip ( service time )  
def generate_vip_service_time(r2):
    if r2<11:
        vip_service_time = 1
    elif r2<41:
        vip_service_time = 2
    elif r2<79:
        vip_service_time = 3
    else:
        vip_service_time = 4
    return vip_service_time
```

```python
# generate the ordinary ( interarrival time ) 

def generate_ord_interarrival_time(r3):
    if r3<10:
        ord_interarrival_time = 0
    elif r3<27:
        ord_interarrival_time = 1
    elif r3<54:
        ord_interarrival_time = 2
    elif r3<74:
        ord_interarrival_time = 3
    elif r3<89:
        ord_interarrival_time = 4
    else:
        ord_interarrival_time = 5
    return ord_interarrival_time
```

```python
#function to generate the ordinary ( service time ) 
def generate_ord_service_time(r4):
    if r4<21:
        ord_service_time = 1
    elif r4<61:
        ord_service_time = 2
    elif r4<89:
        ord_service_time = 3
    else:
        ord_service_time = 4
    return ord_service_time
```

# here we start the code of the simulation 

```python
#here is the number of all customers in the one calendar
calendar_counter = 60
num_of_customer=30
#here this is the number of cases that the whole system will repeated
counter = 1000

#the outer loop of the system ( test cases loop ) 
for i in range(counter):
    #counter to point to the current customer 
    vip_count = 0
    ord_count = 0

    #here i make a clear all lists content when start each new test case 
    vip_random_number_interarrival_time.clear()
    vip_random_number_service_time.clear()
    vip_interarrival_time_queue.clear()
    vip_service_time_queue.clear()
    ord_random_number_interarrival_time.clear() 
    ord_random_number_service_time.clear()
    ord_arrival_time.clear()
    vip_arrival_time.clear()
    ord_service_start_time.clear() 
    vip_service_start_time.clear()
    ord_waiting_time.clear()
    vip_waiting_time.clear()
    vip_comp_time.clear()
    ord_comp_time.clear()
    comp_time.clear()
    t_i_s.clear()
    ord_interarrival_time_queue.clear()
    ord_service_time_queue.clear()
    idle_time_list.clear()
    max_num_stay_in_ord_queue.clear()
    max_num_stay_in_vip_queue.clear()
    
    #for loop to generate the interarrival time and service time for all customers 
    for k in range(num_of_customer):
        r1 = random.randint(100)          # r1 (random generation) for vip 
        r3 = random.randint(100)          # r3 (random generation) for ordinary 


        #this step to generate the interarrival time for both customer type 
        vip_random_number_interarrival_time.append(r1)
        ord_random_number_interarrival_time.append(r3)


        # generate the random number for the interarrival time 
        vip_IAT = generate_vip_interarrival_time(r1)
        ord_IAT = generate_ord_interarrival_time(r3)

        # appending to the list for (interarrival time after generated )
        vip_interarrival_time_queue.append(vip_IAT)
        ord_interarrival_time_queue.append(ord_IAT)

        #get another random number for the service time 
        r2 = random.randint(100)         # r2 for  vip 
        r4 = random.randint(100)         # r4 for  ordinary

        #here appending the random number in its list 
        vip_random_number_service_time.append(r2)
        ord_random_number_service_time.append(r4)

        #here call the functions to generate the service time 
        vip_ST = generate_vip_service_time(r2)
        ord_ST = generate_ord_service_time(r4) 

        #appending the service time after generated
        vip_service_time_queue.append(vip_ST)
        ord_service_time_queue.append(ord_ST)
    

    #for loop to calculate the arrival time for whole customers 
    for j in range(num_of_customer):
        
        #test case for the first customer only 
        if j==0:
            #case that vip less than ordianry 
            if vip_interarrival_time_queue[j] < ord_interarrival_time_queue[j]:
                arrival_time = vip_interarrival_time_queue[j]
                vip_arrival_time.append(arrival_time)
                arrival_time = arrival_time + ord_interarrival_time_queue[j]
                ord_arrival_time.append(arrival_time)

            #case of vip and ordinary are equal 
            elif vip_interarrival_time_queue[j]==ord_interarrival_time_queue[j]:
                arrival_time = vip_interarrival_time_queue[j]
                vip_arrival_time.append(arrival_time)
                ord_arrival_time.append(arrival_time)

            #case that ordiarny is less than vip 
            else:
                arrival_time = ord_interarrival_time_queue[j]
                ord_arrival_time.append(arrival_time)
                arrival_time = arrival_time + vip_interarrival_time_queue[j]
                vip_arrival_time.append(arrival_time)
        
        # test for other customers 
        else:
            
            # check if the interarrival time of the vip less than the interarrival time of ordinary 
            if vip_interarrival_time_queue[j]<ord_interarrival_time_queue[j]:
                arrival_time = vip_interarrival_time_queue[j] + max(vip_arrival_time[j-1],ord_arrival_time[j-1])
                vip_arrival_time.append(arrival_time)
                arrival_time = arrival_time + ord_interarrival_time_queue[j]
                ord_arrival_time.append(arrival_time)
                
            # chekc if the interarrival time of the vip equal to the interarrival tiem of the ordinary 
            elif vip_interarrival_time_queue[j] == ord_interarrival_time_queue[j]:
                arrival_time = vip_interarrival_time_queue[j] + max(vip_arrival_time[j-1],ord_arrival_time[j-1])
                vip_arrival_time.append(arrival_time)
                ord_arrival_time.append(arrival_time)
                
                
            # check if interarrival time of the vip bigger than the interarrival time of ordinary 
            else:
                arrival_time = ord_interarrival_time_queue[j] + max(vip_arrival_time[j-1],ord_arrival_time[j-1])
                ord_arrival_time.append(arrival_time)
                arrival_time = arrival_time + vip_interarrival_time_queue[j]
                vip_arrival_time.append(arrival_time)
                
    #make the operation in the calendar 
    for x in range(calendar_counter):
        # first customer in system
        if x == 0:
            if ord_arrival_time[ord_count] < vip_arrival_time[vip_count]:

                start_time = ord_arrival_time[ord_count]
                idle_time = start_time
                idle_time_list.append(idle_time)
                ord_service_start_time.append(start_time)
                wait_time = start_time - ord_arrival_time[ord_count]
                com_time = start_time + ord_service_time_queue[ord_count]
                T_I_S = com_time - start_time
                ord_waiting_time.append(wait_time)
                comp_time.append(com_time)
                t_i_s.append(T_I_S)
                ord_comp_time.append(com_time)
                ord_count = ord_count+1

            else:
                start_time = vip_arrival_time[vip_count]
                idle_time = start_time
                idle_time_list.append(idle_time)
                vip_service_start_time.append(start_time)
                wait_time = start_time - vip_arrival_time[vip_count]
                com_time = start_time + vip_service_time_queue[vip_count]
                T_I_S = com_time - start_time
                vip_waiting_time.append(wait_time)
                comp_time.append(com_time)
                t_i_s.append(T_I_S)
                vip_comp_time.append(com_time)
                vip_count = vip_count + 1

        #here for the other customers        
        else:
            #here cases of ordinary in system
            if (ord_count<=(len(ord_arrival_time)-1) and vip_count<= (len(vip_arrival_time)-1) and ord_arrival_time[ord_count] < vip_arrival_time[vip_count] and comp_time[x-1] < vip_arrival_time[vip_count]) or (vip_count > (len(vip_arrival_time)-1)):
                if(comp_time[x-1] < ord_arrival_time[ord_count]):
                    start_time = ord_arrival_time[ord_count]
                else:
                    start_time = comp_time[x-1]
                idle_time = start_time - comp_time[x-1]
                idle_time_list.append(idle_time)
                ord_service_start_time.append(start_time)
                wait_time = start_time - ord_arrival_time[ord_count]
                com_time = start_time + ord_service_time_queue[ord_count]
                T_I_S = com_time - start_time
                ord_waiting_time.append(wait_time)
                
                comp_time.append(com_time)
                t_i_s.append(T_I_S)
                ord_comp_time.append(com_time)
                ord_count = ord_count+1

            #case of vip in system
            else:
                if(comp_time[x-1] < vip_arrival_time[vip_count]):
                    start_time = vip_arrival_time[vip_count]
                else:
                    start_time = comp_time[x-1]
                idle_time = start_time - comp_time[x-1]
                idle_time_list.append(idle_time)
                vip_service_start_time.append(start_time)
                wait_time = start_time - vip_arrival_time[vip_count]
                com_time = start_time + vip_service_time_queue[vip_count]
                T_I_S = com_time - start_time
                vip_waiting_time.append(wait_time)
                comp_time.append(com_time)
                t_i_s.append(T_I_S)
                vip_comp_time.append(com_time)
                vip_count = vip_count + 1
                
    # some calculation for the project statistics and analysis
    average_wait_time_ord = round(sum(ord_waiting_time)/len(ord_waiting_time),2)
    average_wait_time_ord_list.append(average_wait_time_ord)

    average_wait_time_vip = round(sum(vip_waiting_time)/len(vip_waiting_time),2)
    avarege_wait_time_vip_list.append(average_wait_time_vip)

    average_idle_time = round(sum(idle_time_list)/len(idle_time_list),2)
    avarege_idle_time_list.append(average_idle_time)
    
    avarege_comp_time_vip = round(sum(vip_comp_time)/len(vip_comp_time),2)
    avarege_comp_time_vip_list.append(avarege_comp_time_vip)

    average_comp_time_ord = round(sum(ord_comp_time)/len(ord_comp_time),2)
    average_comp_time_ord_list.append(average_comp_time_ord)

    avarege_arrival_time_vip = round(sum(vip_arrival_time)/len(vip_arrival_time),2)
    avarege_arrival_time_vip_list.append(avarege_arrival_time_vip)
    
    average_arrival_time_ord = round(sum(ord_arrival_time)/len(ord_arrival_time),2)
    average_arrival_time_ord_list.append(average_arrival_time_ord)
    
    #make the graph 
    if i==100 or i==500 or i==999:
        plt.hist(vip_interarrival_time_queue, density = True , bins = 30)
        plt.ylabel('prob of I_A_T')
        plt.xlabel('Values of I_A_T')
        plt.show()

        plt.hist(vip_arrival_time, density = True , bins = 30)
        plt.ylabel('prob of A_T')
        plt.xlabel('Values of A_T')
        plt.show()

        plt.hist(vip_service_start_time, density = True , bins = 30)
        plt.ylabel('prob of W_T')
        plt.xlabel('Values of W_T')
        plt.show()

        plt.hist(vip_waiting_time, density = True , bins = 30)
        plt.ylabel('prob of S_T')
        plt.xlabel('Values of S_T')
        plt.show()

        plt.hist(vip_service_time_queue, density = True , bins = 30)
        plt.ylabel('prob of C_T')
        plt.xlabel('Values of C_T')
        plt.show()

        plt.hist(vip_comp_time, density = True , bins = 30)
        plt.ylabel('prob of T_I_S')
        plt.xlabel('Values of T_I_S')
        plt.show()

        plt.hist(ord_interarrival_time_queue, density = True , bins = 30)
        plt.ylabel('prob of I_A_T')
        plt.xlabel('Values of I_A_T')
        plt.show()

        plt.hist(ord_arrival_time, density = True , bins = 30)
        plt.ylabel('prob of A_T')
        plt.xlabel('Values of A_T')
        plt.show()

        plt.hist(ord_service_start_time, density = True , bins = 30)
        plt.ylabel('prob of W_T')
        plt.xlabel('Values of W_T')
        plt.show()

        plt.hist(ord_waiting_time, density = True , bins = 30)
        plt.ylabel('prob of S_T')
        plt.xlabel('Values of S_T')
        plt.show()

        plt.hist(ord_service_time_queue, density = True , bins = 30)
        plt.ylabel('prob of C_T')
        plt.xlabel('Values of C_T')
        plt.show()

        plt.hist(ord_comp_time, density = True , bins = 30)
        plt.ylabel('prob of T_I_S')
        plt.xlabel('Values of T_I_S')
        plt.show()
```
